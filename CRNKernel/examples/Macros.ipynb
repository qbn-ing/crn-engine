# CRN Kernel 宏功能演示

这个 Notebook 展示了 CRN Kernel 的宏定义和使用功能。

```crn
%help
```

## 1. 定义宏

```crn
%define ADD(A, B) :(Y)
directive parameters [ k = 0.003 ]
| 10 C
| A ->{k} A + Y
| B ->{k} B + Y
| Y ->{k} C
%end define
```

## 2. 使用宏

```crn
%export expanded
directive simulation { 
  final=1600; 
  plots=[Input1; Input2; Output];
}
directive simulator deterministic

| 10 Input2
| 10 Input1
| 0 Output

%invoke ADD(Input1, Input2) :(Output)
```

## 3. 宏覆盖

```crn
// 重新定义 ADD 宏，使用不同的参数
%define ADD(A, B) :(Y)
directive parameters [ k = 0.01 ]
| A + B ->{k} Y
%end define
```

```crn
%export expanded
directive simulation { 
  final=100; 
  plots=[X1; X2; Result];
}
directive simulator deterministic

| 5 X1
| 5 X2
| 0 Result

%invoke ADD(X1, X2) :(Result)
```

## 6. 调试输出

```crn
%debug on
%define MULTIPLY(X) :(Z)
directive parameters [ k = 0.005 ]
| X ->{k} X + Z
| Z ->{k*0.5} W
%end define
```
