# CRN Kernel 示例 Notebook

这个 Notebook 演示了如何使用 CRN Kernel 进行化学反应网络模拟。

## 简单反应

首先，我们模拟一个简单的化学反应：A + B → C

```crn
%time 100.0
%simulator Stochastic
%seed 42

directive simulation {final=100.0; points=100}
directive stochastic {scale = 500.0}
directive parameters [ k = 0.01 ]

| 10 A
| 10 B
| 0 C

A + B ->{k} C
```

## 连续模拟

下一个单元格会使用上一个单元格的终态作为初始条件。

```crn
%time 50.0

directive simulation {final=50.0; points=100}
directive parameters [ k = 0.05 ]

| 0 D

(* C 分解为 D
   注意：C 的浓度会自动从上一个单元格继承 *)
C ->{k} D
```

## 禁用状态传递

如果需要使用独立的初始条件，可以使用 %preserve false 指令。

```crn
%preserve false
%time 50.0

directive simulation {final=50.0; points=100}
directive stochastic {scale = 500.0}
directive parameters [ k = 0.02 ]

(* 全新的反应系统 *)
| 5 X
| 5 Y
| 0 Z

X + Y ->{k} Z
```
