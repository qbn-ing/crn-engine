# 指令系统演示

这个 Notebook 展示了 CRN Kernel 的各种指令。

所有示例都严格遵循 CRN Engine 项目的原始语法。

```crn
# 设置模拟参数
%time 200.0
%simulator Stochastic
%seed 42

directive sample 0.0,200.0 100
directive plot "A"; "B"; "C"

init "A" 10.0 |
init "B" 10.0 |
init "C" 0.0 |

"A" + "B" ->{0.01} "C"
```

```crn
# 指定要绘制的物种
%plot C, D
%time 100.0

directive sample 0.0,100.0 100

init "D" 0.0 |

"C" ->{0.05} "D"
```

```crn
# 切换模拟器
%simulator Deterministic
%time 100.0

directive sample 0.0,100.0 100
directive plot "D"; "E"

init "E" 0.0 |

"D" ->{0.1} "E"
```

```crn
# 禁用状态传递
%preserve false
%time 50.0

directive sample 0.0,50.0 100
directive plot "X"; "Y"

# 这是一个全新的系统
init "X" 20.0 |
init "Y" 0.0 |

"X" ->{0.2} "Y"
```

```crn
# 重置所有状态
%reset
%time 100.0
%seed 100

directive sample 0.0,100.0 100
directive plot "P"; "Q"; "R"

# 完全从头开始
init "P" 15.0 |
init "Q" 15.0 |
init "R" 0.0 |

"P" + "Q" ->{0.02} "R"
```
