# Lotka-Volterra 捕食者 - 猎物模型

这是一个经典的生态学模型，展示捕食者和猎物种群的周期性振荡。

```crn
%time 200.0
%simulator Deterministic

directive simulation {final=200.0; points=200}
directive parameters [
    c1 = 0.01;
    c2 = 0.02;
    c3 = 0.1;
]

(* 初始种群 *)
init prey 100 |
init predator 10 |
init food 1000 |

(* 猎物繁殖：prey + food -> 2 prey *)
prey + food ->{c1} prey + prey + food |

(* 捕食：prey + predator -> 2 predator *)
prey + predator ->{c2} predator + predator |

(* 捕食者死亡 *)
predator ->{c3}
```

## Brusselator 振荡器

Brusselator 是一个理论化学振荡器模型。

```crn
%preserve false
%time 100.0
%simulator Deterministic

directive simulation {final=100.0; points=200}
directive parameters [
    A = 1.0;
    B = 3.0;
]

(* 初始浓度 *)
init A 1.0 |
init B 3.0 |
init X 1.0 |
init Y 1.0 |

(* Brusselator 反应 *)
->{A} X |
B + X ->{1.0} Y + D |
X + X + Y ->{1.0} X + X + X |
X ->{1.0} E
```
