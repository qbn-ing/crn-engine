# DSD (DNA 链置换) 示例

这个 Notebook 演示了如何使用 Visual DSD (ClassicDSD) 语言编程 DNA 链置换电路。

## 简单的 DNA 链置换反应

这是一个基本的三链 DNA 置换反应。

```crn
%time 7000.0
%simulator Stochastic
%seed 42

directive duration 7000.0 points 100
directive scale 500.0
directive plot <2 3^ 4>(); <4 5^>(); [2 3^ 4]{5^*}()

def S1() = <2 3^ 4>
def S2() = <4 5^>
def Gate() = <1>[2]:<6>[3^ 4]{5^*}
def Product() = [2 3^ 4]{5^*}

(
  13*S1()
| 10*S2()
| 10*Gate()
| 0*Product()
| rxn Gate() + S2() <->{0.00065}{0.004} <1>[2]:<6>[3^ 4]:<4>[5^]
)
```

## Two-Domain Transducer

这是一个经典的二域转导器设计。

```crn
%preserve false
%time 10000.0
%simulator Deterministic

directive duration 10000.0 points 1000
directive compilation infinite
directive plot <t^ x>(); <t^ y>(); <t^ z>()

new t
def T(N,x,y) =
  new a
  (
    N * <t^ a>
  | N * <y t^>
  | N * {t^*}:[x t^]:[a t^]:[a]
  | N * [x]:[t^ y]:[t^ a]:{t^*}
  )

(
  <t^ x>
| T(1,x,y)
)
```

## AND 门电路

使用 DNA 链置换实现的 AND 逻辑门。

```crn
%preserve false
%time 10000.0
%simulator Stochastic
%seed 42

directive duration 10000.0 points 1000
directive plot Input1(); Input2(); Output()
directive scale 10.0

def Input1() = <tb^ b>
def Input2() = <tx^ x>
def Output() = <x to^>
def Join() = {tb^*}[b tx^]:[x to^]

(
  10*Input1()
| 10*Input2()
| 100*Join()
| 0*Output()
)
```
