# DSD (DNA 链置换) 示例

这个 Notebook 演示了如何使用 Visual DSD 语言编程 DNA 链置换电路。

```crn
%time 50.0
%simulator Stochastic
%seed 123

# 定义域
domain a = 10
domain b = 10
domain c = 10
domain tx = 5

# 定义输入链
strand I1: a b
strand I2: b* c*

# 定义门复合物
gate G1: {a*} {b*} tx

# 创建新实例
let x = new I1
let y = new I2
let g = new G1
```
