# CRN Kernel 调试测试

这个 Notebook 用于测试 CRN Kernel 的调试功能。

```crn
%help
```

## 默认情况（调试输出关闭）

```crn
%time 100.0
%simulator Stochastic
%seed 42

directive simulation {final=100.0; points=100}
directive stochastic {scale = 500.0}
directive parameters [ k = 0.01 ]

| 10 A
| 10 B
| 0 C

A + B ->{k} C
```

## 启用调试输出

```crn
%debug on
```

```crn
%time 50.0
%simulator Deterministic

directive simulation {final=50.0; points=50}
directive parameters [ k = 0.1 ]

| 50 X
| 0 Y

X ->{k} Y
```

## 禁用调试输出

```crn
%debug off
```

```crn
%preserve false
%time 50.0
%simulator Deterministic

| 50 X
| 0 Y

X ->{0.1} Y
```
